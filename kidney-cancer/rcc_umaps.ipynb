# RCC UMAPs

Plain `sc.pl.umap` visualizations: categorical labels drawn on the cells, single-gene
expression, and signature scores (from the `scvi_normalized` layer).

Run on a **compute-node kernel** (the integrated file is ~6.8 GB).

```python
import scanpy as sc
import numpy as np
import pandas as pd

sc.settings.set_figure_params(dpi=110, frameon=False)
```

## Load

```python
adata = sc.read_h5ad("Cleaned_Data/integrated.h5ad")
# adata = sc.read_h5ad("Cleaned_Data/myeloid_FINAL_labels.h5ad")   # myeloid subset instead
adata
```

## Categorical labels (on the cells)

```python
sc.pl.umap(adata, color = ['cell type'], frameon = False, legend_loc = "on data")
sc.pl.umap(adata, color = ['leiden'], frameon = False, legend_loc = "on data")
sc.pl.umap(adata, color = ['condition'], frameon = False, legend_loc = "on data")
sc.pl.umap(adata, color = ['Sample'], frameon = False)   # many samples -> keep the legend

# for the myeloid file, use color = ['final_label'] instead of ['cell type']
```

## Single-gene expression

From the `scvi_normalized` layer.

```python
sc.pl.umap(adata, color = ['ATF3', 'JUN', 'FOS', 'CD68', 'LYZ', 'CD3D'],
           frameon = False, layer = 'scvi_normalized', vmax = 10)
```

## Signature scores

Define a gene list, score it, and plot the score.

```python
# AP-1 / immediate-early module
gene_list = ['ATF3', 'JUN', 'JUNB', 'JUND', 'FOS', 'FOSB', 'EGR1', 'NR4A1', 'NR4A2', 'DUSP1']

sc.tl.score_genes(adata, gene_list, score_name = 'AP1_module', use_raw = False, layer = 'scvi_normalized')

sc.pl.umap(adata, color = 'AP1_module', frameon = False, vmax = 10, cmap = 'viridis')
```

```python
# template: swap in any gene list (example: CD8 T signature)
gene_list = [
    'CD8B', 'CCR7', 'CD7', 'CD3D', 'LTB', 'NOSIP', 'CD8A', 'CD3E', 'CD27',
    'IL7R', 'LDLRAP1', 'TCF7', 'LEF1', 'CD3G', 'LDHB', 'RGS10', 'PIK3IP1',
    'OXNAD1', 'ACAP1', 'ATM', 'CD48', 'TRABD2A'
]

sc.tl.score_genes(adata, gene_list, score_name = 'CD8_T_signature', use_raw = False, layer = 'scvi_normalized')

sc.pl.umap(adata, color = 'CD8_T_signature', frameon = False, vmax = 10, cmap = 'viridis')
```
